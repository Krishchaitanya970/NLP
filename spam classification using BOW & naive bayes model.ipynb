```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("spamm.csv")
```

```python
df.head()
```

```python
df.Category.value_counts()
```

```python
df['spam'] = df['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.33, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v= CountVectorizer()

v.fit_transform(X_train)
```

```python
X_train_cv = v.fit_transform(X_train)
X_train_cv
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

model.fit(X_train_cv, y_train)

```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

classification_report(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Testing with an example

Message = [
    'Hey, Mohan. How are you?',
    'Upto 20% discount on parking, Hurry up!'
]

Message_count = v.transform(Message)
model.predict(Message_count)
```
